```python
from sympy import groebner
from sympy.abc import x,y,z
```

```python
f1=x*y+1
f2=y**2-1
```

```python
G=groebner([f1,f2],method='buchberger')
```

```python
G
```

```python
G.contains(x*y**2-x)
G.contains(x*y**2-1)
```
