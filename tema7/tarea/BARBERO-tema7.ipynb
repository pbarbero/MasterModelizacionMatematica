Siguiendo el esquema visto para el caso del laplaciano, escribir un código que ejecute la minimización de la variación total. Aplicarlo a la imagen _boat\_256.png_ corrompida con ruido gaussiano del $5\%$, usando los valores de los parámetros:

$λ = 0.1\text{, }τ = 0.75\text{, }ε = 1 e^{-6}\text{, }\text{num_iter} =20$, si la escala es $I_{255}$,

Mostrar el resultado en una figura.



**Solución:**

Aplicamos un ruido gaussiano del $5\%$ a la imagen,

```python
I = Image.open("boat_256.png")
uc=np.asarray(I,dtype=np.float32)
```

```python
def aplicar_ruido(uc):
    u0=uc+12.*np.random.standard_normal(uc.shape) # 0.04705882352 error
    u0=u0.clip(min=0, max=255)
    return u0
```

```python
u0=aplicar_ruido(uc)
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(uc, cmap='gray', interpolation='nearest')
plt.title("Boat")
plt.subplot(122)
plt.imshow(u0, cmap='gray', interpolation='nearest')
plt.title("Boat con 5% ruido")
```

Definimos la función que aplica la minimización de la variación total,

```python
def variacion_total(u0, lbda, tau, eps):
    u=np.copy(u0)
    #u=u0
    for i in range(0,20):
        GradU=np.gradient(u)
        pupx=GradU[0]
        pupy=GradU[1]
        #Grad_pupx=np.gradient(pupx)
        #Grad_pupy=np.gradient(pupy)
        #p2up2x=Grad_pupx[0]
        #p2up2y=Grad_pupy[1]
        #if norm(GradU) == 0:
        #    lap=(p2up2x+p2up2y)/(sqrt(eps**2+(pupx**2+pupy**2)**2))
        #else:
        #    lap=(p2up2x+p2up2y)/(pupx**2+pupy**2)        
        coefx = pupx / sqrt(eps**2 + pupx**2 + pupy**2)
        coefy = pupy / sqrt(eps**2 + pupx**2 + pupy**2)
        lap = np.gradient(coefx)[0] + np.gradient(coefy)[1]        
        u = u+tau*(lap+lbda*(u0-u))

    u=255*(u-u.min())/(u.max()-u.min())    
    return u
```

```python
u1=np.copy(u0)
restaurada = variacion_total(u1, 0.1, 0.75, 10**(-6))
```

```python
plt.figure(1, figsize=(20,10))
plt.subplot(131)
plt.imshow(uc,cmap='gray',interpolation='nearest')
plt.title('Imagen original')
plt.subplot(132)
plt.imshow(u0,cmap='gray',interpolation='nearest')
plt.title('Imagen con ruido')
plt.subplot(133)
plt.imshow(restaurada,cmap='gray',interpolation='nearest')
plt.title('Restaurada con valores 0.1, 0.75, 1.e-6')
```

- Usar el filtro Yaroslavsky con los parámetros
$h = \rho = 30$, si la escala es $I_{255}$

_Indicación:_ Para la integración del filtro de Yaroslavsky utilizamos la aproximación siguiente,
    
$$ \int_{B(x,\rho)} f(x)\text{d}x = \sum_{x\in B(x,\rho)} f(x) $$    

_Nota:_ El filtro de Yaroslavsky es algo más sencillo de programar que el bilateral, dado que la integración se limita directamente al cuadrado $B(x,\rho)$. En el filtro bilateral, en principio, la integración ha de realizarse sobre toda la imagen. Sin embargo, puesto que la función gaussiana decae a cero muy rápidamente, normalmente se limita el recinto de
integración a un cuadrado de radio $2\rho$.


```python
def yaroslavsky(b0, h, rho):
    b=np.copy(b0)
    for i in range(0,256):
        for j in range(0,256):
            iMin=max(i - rho, 1)
            iMax=min(i + rho, 256)
            jMin=max(j - rho, 1)
            jMax=min(j + rho, 256)
            I=b0[iMin:iMax,jMin:jMax]
            H=exp(-(I-b0[i][j])**2/h**2)
            b[i][j]=sum(H*I)/sum(H)
    return b
```

```python
a_y = yaroslavsky(a2, 30, 30)
```

De este modo, el cuadrado en el que se realiza la integración es de tamaño $61×61$.

Calcular los PSNR’s de las imágenes restauradas respecto la imagen original sin ruido. Obsérvese que PSNR no es una función predefinida, hay que escribirla.

```python
def PSNR(u0, u):
    MSE = ((u0 - u) ** 2).mean(axis=None)
    return 20*log10(u.max()/sqrt(MSE))
```
