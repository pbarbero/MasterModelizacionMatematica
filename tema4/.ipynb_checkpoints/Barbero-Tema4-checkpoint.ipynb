Oirás un mensaje poco claro, por culpa de ruidos con frecuencia alta. Vamos a practicas con bancos de filtros para reconocer el mensaje. Pedimos lo siguiente:

**a.** Pásalo por el banco de Haay y recupéralo sin detalles. Esto debería ser suficiente para reconocer lo que se dice (mediante _wavwrite_ y _subprocess_, como antes); pero aún queda mucho ruido.

```python
ejer6t4=loadmat(ruta+'ejer6t4.mat',squeeze_me=True)
print ejer6t4.keys()
```

```python
y=ejer6t4['y']
wavwrite(y,ruta+'ejer6t4.wav',22050,'float32')
subprocess.Popen([programa,ruta+'ejer6t4.wav'])
```

```python
# página 17 tema4
[h0, h1]=dwt(y,'haar')
```

```python
soloaprox=idwt(h0,0*h1,'haar')
solodet=idwt(0*h0,h1,'haar')
```

```python
figure(3,[20,5])
subplot(1,3,1), plot(y), title(u'Señal de entrada')
subplot(1,3,2), plot(soloaprox), title(u'Aproximación, con el banco de Haar')
subplot(1,3,3), plot(solodet), title(u'Detalles, con el banco de Haar')
```

```python
wavwrite(soloaprox,ruta+'ejer6t4_haar.wav',22050,'float32')
subprocess.Popen([programa,ruta+'ejer6t4_haar.wav'])
```
