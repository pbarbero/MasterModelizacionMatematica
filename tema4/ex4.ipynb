# Ejemplo 5.
Diseñar un filtro pasa banda para señales con frecuencia de muestreo de 8000 Hz con frecuencia de paso de 1000 a 3000 Hz, basado en laventana de Blackman. Usar diferentes
grados del filtro para filtrar un muestreo de tres segundos de duración de la señal analógica x(t) = sen(2π800t) + sen(2π3900t),
intentando eliminar sus dos frecuencias.

```python
N=100 # luego el grado del filtro es 51 = 2N + 1
n=arange(-N,N+1)
w=0.42+0.5*cos(2*pi*n/(2*N+1))+0.08*cos(4*pi*n/(2*N+1)) # ventana de blackman
h=(sin(3*pi/4*n)-sin(pi/4*n))/n/pi # coefs filtro
h[N]=1/2 
f=w*h
F=fftshift(fft(f,8000))
```

```python
hold(True)
plot(arange(-3999,4001),abs(F)), grid(True);
plot([800, 3900],[0.2,0.2],'or')

hold(False)
```

```python
t=linspace(0,3,24000)
x=sin(2*pi*800*t)+sin(2*pi*3900*t)
X=fftshift(fft(x,8000))
y1=lfilter(f,1,x)
Y1=fftshift(fft(y1,8000))
figure(2,[15,8])
subplot(2,1,1), plot(arange(-3999,4001),abs(X),'b'), grid(True)
subplot(2,1,2), plot(arange(-3999,4001),abs(Y1),'r'), grid(True)
```
