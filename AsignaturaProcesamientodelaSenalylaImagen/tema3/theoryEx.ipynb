# Ejercicio página 32

Vamos a tratar con una señal con sólo una frencuencia y ver la importancia que tiene la densidad del muestreo para que sta frecuencua quede bien reflejada o no en las frecuencias de la señal digital. Tomemos
$$f(t) = sin(2pi500t)$$

Su frecuencia es real, $500 Hz$, tiene un periodo de $\frac{1}{500}$. La muestreamos en un periodo de tiempo de $1$ segundo., $[0,1]$, con cinco diferentes muestreos. Así, obtendremos cinco señales digitales finitas distintas, a las cuales aplicaremos un $fft$ y luego un $fft$ añadiendo ceros para ver el espectro de $X(\omega)$

```python
fm=5000
t=linspace(0,1,fm)
x1=sin(2*pi*500*t)
X1=fft(x1)
X11=fft(x1,10000)
```

```python
figure(1,[40,10]);
subplot(2,4,1), plot(abs(X1));
title('Resultado de fft (50000 muestras)');
subplot(2,4,2), plot(linspace(0,pi,5000),abs(X11[0:5000]));
title('Espectro digital');
```

```python
fm=1100
t=linspace(0,1,fm)
x2=sin(2*pi*500*t)
X2=fft(x2)
X22=fft(x2,10000)
subplot(2,1,1), plot(abs(X2));
title('Resultado de fft (1100 muestras)');
subplot(2,1,2), plot(linspace(0,pi,5000),abs(X22[0:5000]));
title('Espectro digital');
```

```python
fm=800
t=linspace(0,1,fm)
x3=sin(2*pi*500*t)
X3=fft(x2)
X33=fft(x2,10000)
subplot(2,1,1), plot(abs(X3));
title('Resultado de fft (1100 muestras)');
subplot(2,1,2), plot(linspace(0,pi,5000),abs(X33[0:5000]));
title('Espectro digital');
```
