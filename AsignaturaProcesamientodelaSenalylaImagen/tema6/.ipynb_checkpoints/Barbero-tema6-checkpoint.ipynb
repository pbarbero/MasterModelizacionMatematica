# Ejercicio 1
Sea la función 1-periódica,

$$ f(x,y) = \dfrac{sen(2\pi x) + cos(4\pi y)}{2 + cos(4\pi x)^3 sin(2\pi y)}$$

Introducir con $meshgrid$ la matriz $A=\big\{f(\frac{j}{151},\frac{k}{151})\big\}^{k=0,\dots,150}_{j=0,\dots,150}$

**a.** Calcular con $fft2$ los coeficientes de Fourier $\hat{f}(m,n)$ para $-75\leq m,n \leq 75$

```python
N=75
Y=X=linspace(0,2*N/(2*N+1), 2*N+1)
[y,x]=meshgrid(Y,X)
f=(sin(2*pi*x) + cos(4*pi*y))/(2+((cos(4*pi*x))**3)*(sin(2*pi*y))) # matriz A
```

# Ejercicio 2
Sea $f$ función 1-periódica tal que

$$ f(x,y) = \chi_{[0,1/4] \times [0,1/4]} (x,y), \text{ } (x,y)\in [0,1) \times [0,1)  $$

**a.** Introducir en Python la matriz de datos,

$$ A = \Big\{ f(\frac{j}{2N+1}, \frac{k}{2N+1}) \Big\}_{j=0,\dots,2N}^{k=0,\dots,2N} $$

para $N=100$. Esta es una matriz que consta de ceros y unos y su introducción en Python no debería suponer mucha dificultad.

```python
N=100

Y=X=linspace(0,2*N/(2*N+1), 2*N+1)
[y,x]=meshgrid(X,Y)

A = 1*((x<=1/4) & (y<=1/4))
```

**b.** A partir de A, calcular, mediante **fft2**, los coeficientes de Fourier aproximados de $f$ . Llamemos a estos coeficientes $a(n, m)$.

Extraer de la matriz total la submatriz $\{ a(n, m)) \}_{n=1,\dots,100}^{m=1,\dots,100} $


```python
a=1/(2*N+1)/(2*N+1)*fft2(A)
a1 = a[1:N+1, 1:N+1]
```

**c.** Calcular los coeficientes de Fourier exactos $\hat{f}(n, m)$ (lo mejor será hacer las sencillas integrales a mano) e introducir en Python la matriz 

 $$ \big\{\hat{f}(n, m))\big\}_{n=1,\dots,100}^{m=1,\dots,100}$$  con la ayuda de _meshgrid_.


$$\hat{f}(n,m) = \int_0^{1/4} e^{-2\pi inx} dx \int_0^{1/4} e^{-2\pi imy} dy =  \Big[ \frac{e^{-2\pi inx}}{-2\pi in}\Big] _{0}^{1/4}  \Big[\frac{e^{-2\pi imy}}{-2\pi im}\Big]_{0}^{1/4} = \frac{e^{-\frac{1}{2}\pi in} - 1}{-2\pi in} \frac{e^{-\frac{1}{2}\pi im} - 1}{-2\pi im} =$$

$$= \frac{1}{4\pi^2 nm}\Big[ cos(\frac{-\pi n}{2}) + i sen(\frac{-\pi n}{2} ) - 1 \Big] \Big[ cos(\frac{-\pi m}{2}) + i sen(\frac{-\pi m}{2}) - 1 \Big]$$

```python
U=V=arange(1,N+1)
[m,n] = meshgrid(V,U)
F=(-1/(4*pi**2*n*m))*(cos(-pi*n/2) + 1j*sin(-pi*n/2) -1)*(cos(-pi*m/2) + 1j*sin(-pi*m/2) - 1)
```

**d.** Calcular el error $E_N = sup\{|\hat{f}(n, m) − a(n, m)| : 1 \leq n, m \leq 100\} $.
($E_N$ depende de $N$ porque los valores aproximados $a(m, n)$ dependen de $N$ ).


```python
EN=abs(F-a1).max()
print EN
```

$E_{100} = 0.00125734038707$ 

**e.** Repetir los apartados anteriores con $N = 200$, $N = 500$ y $N = 1000$. El correspondiente error $E_N$ debería mejorar, pero no lo hace de forma espectacular. ¿Por qué?

Hay que cambiar también la cantidad de valores que extremos para la submatriz de $a$.

$ E_{200} = 0.000628896876367$

$ E_{500} = 0.000251611545801$

$ E_{1000} = 0.000125814435927$

Por el lema de Riemann-Lebesgue, tenemos que el límite de $\hat{f}$ cuando $(|n|,|m|)\rightarrow (0,0)$ es $0$. Por lo que la diferencia de errores no va a ser muy grande por mucho que aumentemos la $N$.

En el archivo _ejer3t6.mat_ tenemos una matriz $g$ de dimensiones $501\times 501$ que corresponde a un muestreo

de una señal 1-periódica $f$.

**a.** ¿Sabrías deducir que $f$ es una señal de banda limitada?

```python
from scipy.io import loadmat
import subprocess
ruta = "/home/pilar/Master/procesamiento/tema6/"
```

```python
ejer3t6=loadmat(ruta+'ejer3t6.mat',squeeze_me=True)
g=ejer3t6['g']
```

Vamos a procesar su transformada de Fourier, y veremos si cumple que una señal de banda limitada, su transformada de Fourier tiene que cumplir, para $c(n,m)\in \mathbb{C}$

$$ \hat{f} = c(n,m),\text{ }si\text{ } |n|\leq N \text{ } y \text{ } |m|\leq M $$

y en otro caso, vale $0$

```python
N=250
Y=X=linspace(0,2*N/(2*N+1), 2*N+1)
[y,x]=meshgrid(X,Y)
G=1/(2*N+1)/(2*N+1)*fft2(g)
G1=G[0:N+1, 0:N+1]
G2=G[0:N+1, N+1:2*N+1]
G3=G[N+1:2*N+1, 0:N+1]
G4=G[N+1:2*N+1, N+1:2*N+1]

Gcen=vstack((hstack((G4,G3)),hstack((G2,G1))))
```

```python
figure(1)
fig=figure(1).add_subplot(111,projection='3d')
fig.plot_surface(x,y,abs(Gcen),rstride=20,cstride=20,cmap=cm.jet);
```

Podemos observar que la señal es de banda limitada porque su transformada de Fourier toma valores $0$ un cuadrado donde estan los coeficientes $c(n,m)$ que posteriormente hallaremos. A simple vista podemos observar que se trata de un cuadrado $40\times 40$, y su valores son aproximadamente $0.25$

**b.** ¿Sabrías demostrar que $f$ es de la forma 

$$ f(x,y) = \sum_{n=1}^{p} \sum_{m=1}^{p} c(n,m) sen(2\pi nx) cos(2\pi my)$$

encontrando el valor $p$ y los coeficientes $c(n,m)$?

Como $f$ es de banda limitada, sabemos que la señal $f$ es de la forma $$\sum_{n=-N}^{N} \sum_{m=-M}^{M} c(n,m)e^{2\pi inx}e^{2\pi imy}$$

Vamos a trabajar con los coeficientes de Fourier, en el apartado anterior hemos probado que la parte real de los coeficientes es prácticamente nula.

```python
imshow(real(Gcen), interpolation='nearest',cmap=cm.jet), colorbar();
```

Tomamos ólo la parte imaginaria.

```python
imshow(imag(Gcen), interpolation='nearest',cmap=cm.jet), colorbar();
```

En la mitad superior toma valores $0.25i$ y en la parte inferior $-0.25i$. Los valores que toma en las rectas $(n,0)$ y $(0,m)$ son prácticamente nulos.

```python
print G[0,0]
print G[:,0].max(), G[:,0].min()
print G[0,:].max(), G[0,:].min()
```

Así podemos deducir que $f$ es de la forma, 

$$\sum_{n=-N}^{-1} \sum_{m=-M}^{-1} -0.25i e^{2\pi inx}e^{2\pi imy} + \sum_{n=1}^{N} \sum_{m=-M}^{-1} 0.25i e^{2\pi inx}e^{2\pi imy} + \sum_{n=-N}^{-1} \sum_{m=1}^{M} -0.25i e^{2\pi inx}e^{2\pi imy} + \sum_{n=1}^{N} \sum_{m=1}^{M} 0.25i e^{2\pi inx}e^{2\pi imy} = $$



$$ =  \sum_{n=1}^{N} \sum_{m=1}^{M} \Big[ -0.25ie^{-2\pi inx}e^{-2\pi imy} + 0.25ie^{2\pi inx}e^{-2\pi imy} -0.25ie^{-2\pi inx}e^{2\pi imy} + 0.25ie^{2\pi inx}e^{2\pi imy} \Big]$$

$$ = \sum_{n=1}^{N} \sum_{m=1}^{M} 0.25 i \Big[ - (\cos(2\pi nx) - i \sin(2\pi nx)) (\cos(2\pi my) - i \sin(2\pi my)) + $$

$$ (\cos(2\pi nx) + i \sin(2\pi nx)) (\cos(2\pi my) - i \sin(2\pi my)) + $$

$$ - (\cos(2\pi nx) - i \sin(2\pi nx)) (\cos(2\pi my) - i \sin(2\pi my)) + $$

$$ (\cos(2\pi nx) + i \sin(2\pi nx)) (\cos(2\pi my) + i \sin(2\pi my)) \Big] $$

Haciendo los productos y las cuentas, vemos que muchos miembros se cancelan, quedando al final,

$$\sum_{n=1}^{N} \sum_{m=1}^{M} 0.25 i (4 i \sin(2\pi nx) \cos(2\pi my)) = \sum_{n=1}^{N} \sum_{m=1}^{M} -\sin(2\pi nx) \cos(2\pi my)$$

Así $c(n,m) = 1 \text{, } \forall (n,m) \in \mathbb{Z} \times \mathbb{Z}$ y $p = 40$ como deduciremos en el apartado **c.**

# Ejercicio 6
Observa la siguiente imagen. Es la variable $a$, contenida en el archivo _ejer6t6.mat_. Es un ejemplo simplificado de un hecho que hemos comentado ya varias veces: parece que es una señal regular, pero al tratarla como periódica ya no lo es, pues hay un gran salto entre la línea de arriba y la de abajo.

```python
ejer6t6=loadmat(ruta+'ejer6t6.mat',squeeze_me=True)
a = ejer6t6['a']
imshow(a, interpolation='nearest',cmap=cm.gray);
len(a), len(a[0])
```

Hemos ido viendo en los ejemplos y problemas precedentes que este hecho representa un gran inconveniente al querer tratar la imagen con las técnicas de frecuencias, propias de señales periódicas. De hecho, el problema de los bordes de una fotografía representa un problema con cualquier otra técnica que se use. Por ejemplo, si queremos ampliar la foto al doble de filas y columnas (es decir, cuatro veces) o más, aparecerá el fenómeno de Gibbs. ¿Cómo se podría evitar este problema de discontinuidad en los bordes para que la ampliación no refleje ninguna deficiencia debida al fenómeno de Gibbs?

Tratándola como periódica,

```python
FOO = vstack(( a, a))
```

```python
figure(2,[10,7])
imshow(FOO,cmap=cm.gray,interpolation='nearest',aspect='equal');
```

Vamos a ampliarla para observar la aparición del fenómeno de Gibbs,

```python
aF=1/256/256*fft2(a)

aF1=aF[0:128,0:128]
aF2=aF[0:128,128:256]
aF3=aF[128:256,0:128]
aF4=aF[128:256,128:256]
```

```python
AMP=vstack(( 
        hstack((aF1,zeros((128,128)),aF2)),
        zeros((128,384)),
        hstack((aF3,zeros((128,128)),aF4))
    ))
```

```python
amp=256*256*ifft2(AMP)
```

```python
figure(1,[20,7])
subplot(1,2,1),imshow(a,cmap=cm.gray,interpolation='nearest',aspect='equal'), colorbar();
subplot(1,2,2),imshow(abs(amp),cmap=cm.gray,interpolation='nearest',aspect='equal'),colorbar();
```

Vamos a probar a evitar el fenómeno de Gibbs, tomando el negativo de la imagen y concateándola dos veces en vertical.

```python
nueva=vstack(( 256-a, a ))
```

```python
figure(1,[20,7])
imshow(nueva,cmap=cm.gray,interpolation='nearest',aspect='equal'),colorbar();
```

Vamos a ampliarla al doble de filas y columnas,

```python
nuevaF=1/256/512*fft2(nueva)
nuevaF1=nuevaF[0:257, 0:129]
nuevaF2=nuevaF[0:257, 129:256]
nuevaF3=nuevaF[257:512, 0:129]
nuevaF4=nuevaF[257:512, 129:256]

nuevaMP=vstack((
                hstack((nuevaF1,zeros((257,256)),nuevaF2)),
                zeros((512,512)),
                hstack((nuevaF3,zeros((255,256)),nuevaF4))
              ))
```

```python
nueva_mp=1024*512*real(ifft2(nuevaMP))
figure(1,[20,7])
subplot(1,2,1),imshow(nueva_mp,cmap=cm.gray,interpolation='nearest',aspect='equal'), colorbar();
subplot(1,2,2),imshow(nueva,cmap=cm.gray,interpolation='nearest',aspect='equal'), colorbar();
```

Ahora nos quedamos con la parte inferior de la imagen nueva remuestreada y obtendremos la imagen original,

```python
img_final = nueva_mp[513:1024, 0:512]

figure(1,[7,7])
imshow(img_final,cmap=cm.gray,interpolation='nearest',aspect='equal'), colorbar();
```
