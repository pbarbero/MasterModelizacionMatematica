```python
# Ejercicio  mitad negro mitad blanco
C=vstack((zeros((200,400)),ones((200,400))))
imshow(C,cmap=cm.gray,interpolation='nearest',aspect='equal')
```

```python
cfC=1/400/400*fft2(C)
```

```python
n=m=300
```

```python
cfC1=cfC[0:201, 0:201]
cfC2=cfC[0:201, 201:400]
cfC3=cfC[201:400, 0:201]
cfC4=cfC[201:400, 201:400]
```

```python
AMP=vstack((
hstack((cfC1,zeros((201,m)),cfC2)),
zeros((n,400+m)),
hstack((cfC3,zeros((199,m)),cfC4))))
```

```python
amp=(400+n)*(400+m)*ifft2(AMP)
amp=real(amp)
```

```python
figure(1,[14,7])
subplot(1,2,1),imshow(C,cmap=cm.gray,interpolation='nearest',aspect='equal')
subplot(1,2,2),imshow(amp,cmap=cm.gray,interpolation='nearest',aspect='equal')
```

```python
# incorporación de imágenes a python
ruta='/home/pilar/Master/procesamiento/tema6'
daub=imread(ruta+'daubechies.jpg')
```

```python
print daub.shape, daub.dtype # imagen 326 x 228 enteros del 0 a 2**8
```

```python
daub=daub.astype(float)
print daub.shape, daub.dtype # aritmetica habitualfigure(1)
```

```python
figure(1)
imshow(daub,cmap=cm.gray,interpolation='nearest',aspect='equal')
```

```python
dr=dstack((
daub/255,
zeros((326,228)),
zeros((326,228))
))
imshow(dr,interpolation='nearest',aspect='equal')
```

```python
imshow(255-daub,cmap=cm.gray,interpolation='nearest',aspect='equal')
```

```python
DB=fft2(daub)/326/228 # fourier2
X=linspace(0,325,326)
Y=linspace(0,227,228)
[y,x]=meshgrid(Y,X)
figure(1)
fig=figure(1).add_subplot(111,projection='3d')
fig.plot_surface(x,y,abs(DB))
```

```python
# compresion de la fotografia
DB1=DB[0:80,0:60]
DB2=DB[0:80,169:228]
DB3=DB[247:326,0:60]
DB4=DB[247:326,169:228]
REC=vstack((hstack((DB1,DB2)),hstack((DB3,DB4))))
```

```python
DB1=REC[0:80,0:60]
DB2=REC[0:80,60:119]
DB3=REC[80:159,0:60]
DB4=REC[80:159,60:119]
AMPREC=vstack((
hstack((DB1,zeros((80,109)),DB2)),
zeros((167,228)),
hstack((DB3,zeros((79,109)),DB4))
))

rec=ifft2(AMPREC)*326*228

rec=real(rec)
```

```python
figure(1)
subplot(1,3,1)
imshow(daub,cmap=cm.gray,interpolation='nearest',aspect='equal')
subplot(1,3,2)
imshow(rec, cmap=cm.gray, interpolation='nearest', aspect='equal')
print daub.min(), daub.max()
print rec.min(), rec.max()
```

```python
# eliminacion de ruidos en una fotografia
from __future__ import division
from numpy.fft import *
from mpl_toolkits.mplot3d import Axes3D
from scipy.io import loadmat
```

```python
foto=loadmat(ruta+'fotoruido.mat')
print foto.keys()
```

```python
fotoruido=foto['fotoruido']
```

```python
print fotoruido.shape, fotoruido.dtype
```

```python
figure(1)
imshow(fotoruido,cmap=cm.gray,interpolation='nearest', aspect='equal')
```

```python
C=fft2(fotoruido)/326/228
X=linspace(0,325,326)
Y=linspace(0,227,228)
[y,x]=meshgrid(Y,X)
fig=figure(2).add_subplot(111,projection='3d', elev=10)
fig.plot_surface(x,y,abs(C),rstride=10,cstride=3)
```

```python
# aplicamos filtro
filtro=abs(C)<0.4
```

```python
REC=filtro*C
rec=ifft2(REC)*326*228
rec=real(rec)
figure(3)
imshow(rec,cmap=cm.gray,interpolation='nearest', aspect='equal')
```
