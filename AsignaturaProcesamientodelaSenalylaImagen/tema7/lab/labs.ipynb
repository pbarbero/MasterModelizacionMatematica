```python
I = Image.open("./images/cameraman_256.tif")
uc=np.asarray(I,dtype=np.float32)
u0=uc+12.*np.random.standard_normal(uc.shape)
u0=u0.clip(min=0,max=255)
plt.figure()
plt.subplot(121)
plt.imshow(uc,cmap='gray',interpolation='nearest')
plt.title("Cameraman")
plt.subplot(122)
plt.imshow(u0,cmap='gray',interpolation='nearest')
plt.title('Cameraman con ruido')
plt.show()

```

```python
[m,n]=np.shape(u0)
UC=np.fft.fftshift(np.fft.fft2(uc))/(m*n)
U0=np.fft.fftshift(np.fft.fft2(u0))/(m*n)
PC=np.abs(UC)**2
P0=np.abs(U0)**2
PR=P0/PC
```

```python
plt.hold(True)
plt.plot(np.arange(0,m),uc[:,n/2],color='blue')
plt.plot(np.arange(0,m),u0[:,n/2],color='red')
plt.title(r'Intensidad de grises en una secci$\’{\rm o}$n de las im$\’{\rm a}$genes limpia y con')
```

```python
plt.figure()
plt.imshow(np.log(1+PR[m/2:-1,n/2:-1]),origin='lower')
plt.title(r'Comparaci$\’{\rm o}$n del espectro de potencias de las im$\'{\rm a}$genes')
```

```python
plt.show()
```

```python
duc=np.gradient(uc[:,n/2],1)
du0=np.gradient(u0[:,n/2],1)
```

```python
plt.hold(True)
plt.plot(np.arange(0,m),duc,color='blue')
plt.plot(np.arange(0,m),du0,color='red')
plt.show()
```

```python
lc=np.sum(np.abs(duc))
# longitud yendo por la imagen limpia
l0=np.sum(np.abs(du0))
# longitud yendo por la imagen con ruido
print(lc,l0)
```

### Ejemplo

```python
tau=0.03
lbda=1
u=u0
for i in range (0,20):
    GradU=np.gradient(u)
    pupx=GradU[0]
    pupy=GradU[1]
    Grad_pupx=np.gradient(pupx)
    Grad_pupy=np.gradient(pupy)
    p2up2x=Grad_pupx[0]
    p2up2y=Grad_pupy[1]
    lap=p2up2x+p2up2y
    u=u+tau*(lap+lbda*(u0-u))
```

```python
u=255*(u-u.min())/(u.max()-u.min())
plt.figure()
plt.subplot(121)
plt.imshow(u0,cmap='gray',interpolation='nearest')
plt.title("Original con ruido")
plt.subplot(122)
plt.imshow(u,cmap='gray',interpolation='nearest')
plt.title("Restaurada")
```
