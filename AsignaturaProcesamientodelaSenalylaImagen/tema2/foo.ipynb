# Ejercicio 1

Sean las funciones $1$-periodicas tales que 
$p_1(t) = t^3 - \frac{3}{2} t^2 + \frac{1}{2} t, p_2(t) = t^3 + \frac{3}{2} t^2 + \frac{1}{2} t$

a. Calcular, mediante \textbf{fft}, los coeficientes de Fourier aproximados con los muestreos 

$fi = \Big\{ p_i \big( \dfrac{k}{2M+1} \big) \Big\}$
con $M = 10000$
.

```python
%pylab inline
from __future__ import division
from numpy.fft import *

close('all')
```

```python
M=10000
t=linspace(0,2*M/(2*M+1),2*M+1)

f1=t**3 - 3/2*t**2 + 1/2*t
f2=t**3 + 3/2*t**2 + 1/2*t
```

```python
figure(1,[20.5,8.5])
subplot(2,2,1),plot(t,f1),grid(True)
title(u'La función p1');
subplot(2,2,2),plot(t,f2),grid(True)
title(u'La función p2');
```

```python
F1=1/(2*M+1)*fft(f1)
F2=1/(2*M+1)*fft(f2)
```

```python
subplot(2,2,2), plot(abs(F1),color='red',linewidth=1), grid(True)
title(u'Coefs Fourier f1');
```

```python
subplot(2,2,2), plot(abs(F2),color='red',linewidth=1), grid(True)
title(u'Coefs Fourier f2');
```

```python
FF1=concatenate((F1[M+1:2*M+1],F1[0:M+1]))
FF2=concatenate((F2[M+1:2*M+1],F2[0:M+1]))
```

```python
s=linspace(-10000,10000,20001)
subplot(2,2,3), plot(s,abs(FF1),color='red',linewidth=2), grid(True)
title('Coeficientes de Fourier f1')
```

```python
subplot(2,2,3), plot(s,abs(FF2),color='red',linewidth=2), grid(True)
title('Coeficientes de Fourier f2')
```

```python
Y1=F1[0:21]
Y2=F2[0:21]
```

```python
figure(1,[20.3,8.5])
subplot(2,2,1), plot(abs(Y1),'o'),grid(True)
title('Coeficientes de Fourier f1, abs, desde orden 0 a 20');
subplot(2,2,2), plot(abs(Y2),'o'),grid(True)
title('Coeficientes de Fourier f2, abs, desde orden 0 a 20');
```
