```python
a=1/(2*N+1)/(2*N+1)*fft2(A)
a1 = a[1:N+1, 1:N+1]
```

 $$ \big\{\hat{f}(n, m))\big\}_{n=1,\dots,100}^{m=1,\dots,100}$$  con la ayuda de _meshgrid_.


$$\hat{f}(n,m) = \int_0^{1/4} e^{-2\pi inx} dx \int_0^{1/4} e^{-2\pi imy} dy =  \Big[ \frac{e^{-2\pi inx}}{-2\pi in}\Big] _{0}^{1/4}  \Big[\frac{e^{-2\pi imy}}{-2\pi im}\Big]_{0}^{1/4} = \frac{e^{-\frac{1}{2}\pi in} - 1}{-2\pi in} \frac{e^{-\frac{1}{2}\pi im} - 1}{-2\pi im} $$

```python
U=V=arange(1,N+1)
[m,n] = meshgrid(V,U)
F=(-1/(4*pi**2*n*m))*(cos(pi*n/2) + 1j*sin(pi*n/2))*(cos(pi*m/2) + 1j*sin(pi*m/2))
```

($E_N$ depende de $N$ porque los valores aproximados $a(m, n)$ dependen de $N$ ).
